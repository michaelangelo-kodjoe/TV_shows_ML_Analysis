## RETRIEVE DATA FROM TMDB API

```python
run_code = input("Enter 'run' to execute API requests.   ")

if run_code == "run":

    #Create an empty list to hold the movie data.
    tv_data = []
    # Print the beginning of the logging.
    print("Beginning Data Retrieval     ")
    print("-----------------------------")

    # # Create counters.
    # record_count = 1
    # set_count = 1
    # Loop through all the cities in the list.
    for i in range(len(tv_shows)):

    #     # Group cities in sets of 50 for logging purposes.
    #     if (i % 50 == 0 and i >= 50):
    #         set_count += 1
    #         record_count = 1
    #         time.sleep(60)

        # Create endpoint URL with each city.
        tv_show = tv_shows[i]
        tv_show_url = f"https://api.themoviedb.org/3/search/tv?api_key={api_key}&query={tv_show.replace(' ','+')}"

        # Log the URL, record, and set numbers and the city.
        print(f"Processing Record {i} | {tv_show}")

        # Run an API request for each of the cities.

        try:
            # Parse the JSON and retrieve data.
            show_info = requests.get(tv_show_url).json()
            # Parse out the needed data.
            genre = show_info['results'][0]['genre_ids']
            show_id = show_info['results'][0]['id']
            name = show_info['results'][0]['name']
            overview_txt = show_info['results'][0]['overview']
            vote_avg = show_info['results'][0]['vote_average']
            vote_count = show_info['results'][0]['vote_count']
            # Append the city information into city_data list.
            tv_data.append({"Show_ID_tmdb": show_id,
                              "name_tmdb": name,
                              "Genre": genre,
                              "Overview_Text_tvdb": overview_txt,
                              "Vote_Average_tmdb": vote_avg,
                              "Vote_Count_tmdb": vote_count})

    # If an error is experienced, skip the tv show.
        except:
            print("TV show not found. Skipping...")
            pass

    # Indicate that Data Loading is complete.
    print("-----------------------------")
    print("Data Retrieval Complete      ")
    print("-----------------------------")

run_code = 0
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
tv_data_df = pd.DataFrame(tv_data)

tv_data_df.to_csv('show_data_retrieved.csv')
len(tv_data_df)
```

```python
#Find shows with same title. 

df = tv_data_df.groupby('name_tmdb').count()

duplicate_titles = list(df[df.Genre >1].index)

get_years_for_these_shows = []

for title in duplicate_titles:
    check_entries = tv_data_df[tv_data_df.name_tmdb == title]
    check_descs = list(check_entries.Overview_Text_tvdb)
    if check_descs[0] == check_descs[1]:
        ##Drop a row of check[1] from tv_data
        tv_data_df.drop(check_entries.index[1], inplace = True)
    else:
        # Get Year for Check[1] and Change name of check[1]
        get_years_for_these_shows.append(check_entries.reset_index().Show_ID_tmdb[1])
        
```
