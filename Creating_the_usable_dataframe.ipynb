```python
#Import Libraries
        
import pandas as pd
from nrclex import NRCLex
```

```python
#Load CSV.. a text to columns process was already performed to separate the season field. 

file_to_load = 'selected_reviews.csv'

df = pd.read_csv(file_to_load)
df.info()
```

```python
#Drop Rows with blank review content
df=df[df.review_content.notnull()]

df.isnull().sum()
```

```python
#Fill null values in season column with 0
#These programs are tv_specials and mini series. Consider dropping them from the dataset entirely.
df['season'].fillna(0, inplace = True)
df['title'][df['season']==0].unique()

```

# Applying NRCLex to Data

```python
#Choose a sample dataframe that will and a list of sample reviews for the same data. 
sample_size = 1000
sample_df = df.head(sample_size)
reviews = df['review_content'][0:sample_size].tolist()
reviews
```

```python
#Declare variable for emotions scores
emotions_scores = []
```

```python
#loop to genearate a dataframe of emotion scores of reviews using NRC Library.

for review in reviews:
    text_object = NRCLex(review)
    data = text_object.raw_emotion_scores
    af = text_object.affect_frequencies
    
    emotions_scores.append(af)

    
print(emotions_scores)    
    
    



```

```python
#Create a Datafrae from the emotions_scores list of dictionaries. 
emotions_df = pd.DataFrame(emotions_scores)
emotions_df
```

```python
#Merge the emotions Scores DataFrame with the Reviews DataFrame

df_merged = sample_df.merge( emotions_df, left_index = True, right_index = True)

df_merged
```

```python
#Export Merged DataFrame to CSV

df_merged.to_csv('tv_emotions_data.csv')
```
