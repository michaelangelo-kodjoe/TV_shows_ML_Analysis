```python
# Importing libraries
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
import sqlite3
```

```python
# # Create a connection to database
engine = create_engine("sqlite:///emotion_ML.sqlite")
```

```python
# Extracting tables from sqlite database and creating a join
joined_df = pd.read_sql_query('''SELECT * 
                    FROM emotion_scores as es
                    JOIN critic_meta_score as cm
                    JOIN review_data as rd
                    ON es.'index' = cm.'index' AND cm.'index' = rd.'index' 
                        ;''', engine)
```
